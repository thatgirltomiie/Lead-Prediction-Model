### Objective: 
Develop a predictive lead scoring model to prioritize leads based on their likelihood of conversion, enabling the sales team to focus efforts on high-value leads.

### Goal:
Create a model that predicts the probability of lead conversion (e.g., "High", "Medium", "Low").

Help the sales team efficiently allocate resources to the most promising leads.

## SPLIT DATA

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(columns=['lead_converted'])
y = data['lead_converted']

X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size = 0.2, stratify=y, random_state = 42 )
print(f"X_train: {X_train.shape}")
print(f"y_train: {y_train.shape}")
print(f"X_test: {X_test.shape}")
print(f"y_test: {y_test.shape}")
```

## MODEL DEVELOPMENT

```python
from xgboost import XGBClassifier 
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Check target classes
y.value_counts()
```

```python
# negative: positive
ratio = 1189/134
ratio
```

```python
# copy data
ranked_leads = X_test.copy()
```

```python
# function to categorize leads
def categorize_lead(score):
    if score >= 0.55:
        return 'High'
    elif score >= 0.20:
        return 'Medium'
    else:
        return 'Low'
```

```python
# initialize and train the models
classifiers = [[LogisticRegression(random_state=42, class_weight='balanced'), 'Logistic Regression'],
               [SGDClassifier(random_state=42, class_weight='balanced', loss='log_loss'), 'SGD Classifiers'],
               [DecisionTreeClassifier(random_state=42, class_weight='balanced'), 'Decision Tree'],
               [RandomForestClassifier(random_state=42, class_weight='balanced'), 'Random Forest'],
               [XGBClassifier(random_state=42, scale_pos_weight=ratio), 'XG Boost']
]

for classifier in classifiers:
    model = classifier[0]
    model_name = classifier[1]
    print(f"<-------{model_name}------>")

    model.fit(X_train, y_train)
   
    lead_conversion_score = model.predict_proba(X_test)[:, 1]
    ranked_leads['lead_conversion_score'] = lead_conversion_score

    # Apply the function to create a new column
    ranked_leads['lead_rank'] = ranked_leads['lead_conversion_score'].apply(categorize_lead)

    predictions = model.predict(X_test)

    report = classification_report(y_test, predictions)
    matrix = confusion_matrix(y_test, predictions)
    print(f"report: \n", report)
    print("matrix: \n", matrix)
    sns.heatmap(matrix, annot = True, fmt = "g", cmap= 'autumn')
    plt.ylabel("Actual predictions")
    plt.xlabel("Model predictions")
    plt.show()

    # Categorize predictions
    actual_predictions = pd.Series(y_test, index=X_test.index).map({1: 'lead_converted', 0: 'lead_not_converted'})

    # Visualize
    cm = pd.crosstab(actual_predictions, ranked_leads['lead_rank'], 
            rownames=['Actual predictions'], colnames=['Model predictions'])
    print(cm)

    sns.heatmap(cm, annot=True, fmt='g', cmap= 'autumn')
    plt.ylabel("Actual predictions")
    plt.xlabel("Model predictions")
    plt.show()
```

## XGBoost is the best
- It does a better job at correctly identifying people who are actually going to convert and not at the expense of wrongly predicting a lot

## FINE-TUNING

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

```python
model = XGBClassifier(
    random_state=42,
    scale_pos_weight=ratio,
    use_label_encoder=False,
    eval_metric='logloss'
)
```

```python
# use of skf for cv since class is unbalanced
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model, X_train, y_train, cv=skf, scoring='f1')

print("Stratified f1 scores:", scores)
```

```python
# More conservative hyperparameter range
grid_params = {
    'max_depth': [2, 3, 4],              
    'learning_rate': [0.01, 0.05],       
    'n_estimators': [50, 100],           
    'subsample': [0.8, 1.0],             
    'colsample_bytree': [0.8, 1.0]
}

grid_search = GridSearchCV(
    estimator=model,
    param_grid=grid_params,
    scoring='f1', 
    cv=skf,
    verbose=1,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
# Output best parameters
print("Best parameters from GridSearchCV:")
print(grid_search.best_params_)
```

```python
# Retrieve best estimator
best_model = grid_search.best_estimator_

# Get predicted probabilities for positive class
lead_conversion_score = best_model.predict_proba(X_test)[:, 1]

# Track model predictions on test data
ranked_leads['lead_conversion_score'] = lead_conversion_score

# Apply the function to create a new column
ranked_leads['lead_rank'] = ranked_leads['lead_conversion_score'].apply(categorize_lead)
```

```python
# make conversion predictions
predictions = best_model.predict(X_test)
```

```python
# evaluate 
report = classification_report(y_test, predictions)
matrix = confusion_matrix(y_test, predictions)
print(f"report: \n", report)
print("matrix: \n", matrix)
sns.heatmap(matrix, annot = True, fmt = "g", cmap = 'autumn')
plt.ylabel("Actual predictions")
plt.xlabel("Model predictions")
plt.show()

# Categorize predictions
actual_predictions = pd.Series(y_test, index=X_test.index).map({1: 'Converted', 0: 'Not Converted'})

# Visualize
cm = pd.crosstab(actual_predictions, ranked_leads['lead_rank'], 
            rownames=['Actual predictions'], colnames=['Model predictions'])
print(cm)

sns.heatmap(cm, annot=True, fmt='g', cmap = 'autumn')
plt.ylabel("Actual predictions")
plt.xlabel("Model predictions")
plt.show()
```
